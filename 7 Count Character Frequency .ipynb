```python
text = "python"
char_count = {}
for char in text:
    if char in char_count:
        char_count[char] += 1 
        print(f"  '{char}' found again, count: {char_count[char]}")
    else:
        char_count[char] = 1 
        print(f"  '{char}' found first time, count: 1")

print(f"\nWord: '{text}'")
print(f"Character frequency: {char_count}")


print("\n--- Method 2: Using Get Method ---")
char_count2 = {}
for char in text:
    char_count2[char] = char_count2.get(char, 0) + 1

print(f"Character frequency: {char_count2}")

print("\n--- Method 3: Using Counter (Built-in) ---")
from collections import Counter
char_count3 = Counter(text)
print(f"Character frequency: {dict(char_count3)}")


print("\n--- Detailed Breakdown ---")
for char, count in sorted(char_count.items()):
    bar = "█" * count
    print(f"  '{char}': {count} time(s) {bar}")


print("\n" + "=" * 60)
print("Real-Life Application: Password Strength Checker")
print("=" * 60)
password = "python123"
char_freq = {}


digits = 0
lowercase = 0
uppercase = 0
special = 0

for char in password:
    char_freq[char] = char_freq.get(char, 0) + 1
    
    if char.isdigit():
        digits += 1
    elif char.islower():
        lowercase += 1
    elif char.isupper():
        uppercase += 1
    else:
        special += 1

print(f"Password: '{password}'")
print(f"Character frequency: {char_freq}")
print(f"\nPassword Analysis:")
print(f"  Lowercase letters: {lowercase}")
print(f"  Uppercase letters: {uppercase}")
print(f"  Digits: {digits}")
print(f"  Special characters: {special}")

# Strength check
strength = "Weak"
if len(password) >= 8 and digits > 0 and lowercase > 0 and uppercase > 0:
    strength = "Strong"
elif len(password) >= 6 and (digits > 0 or lowercase > 0):
    strength = "Medium"

print(f"\nPassword Strength: {strength}")

most_common = max(char_freq, key=char_freq.get)
print(f"Most repeated character: '{most_common}' ({char_freq[most_common]} times)")
```
