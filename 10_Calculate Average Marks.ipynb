```python
marks = [85, 90, 60, 75]

print("\n--- Method 1: Using Built-in Functions ---")
total_marks = sum(marks)
number_of_subjects = len(marks)
average = total_marks / number_of_subjects

print(f"Marks in each subject: {marks}")
print(f"Total marks: {total_marks}")
print(f"Number of subjects: {number_of_subjects}")
print(f"Average Marks: {average}")


print("\n--- Method 2: Manual Calculation ---")
total = 0
for mark in marks:
    total += mark
    print(f"  Adding {mark}, Running total: {total}")

manual_average = total / len(marks)
print(f"Average: {manual_average}")


print("\n" + "=" * 60)
print("Grade Calculation")
print("=" * 60)

if average >= 90:
    grade = "A+"
    feedback = "Outstanding! Excellent performance!"
elif average >= 80:
    grade = "A"
    feedback = "Very Good! Keep it up!"
elif average >= 70:
    grade = "B"
    feedback = "Good! You're doing well!"
elif average >= 60:
    grade = "C"
    feedback = "Satisfactory. Room for improvement!"
elif average >= 50:
    grade = "D"
    feedback = "Pass. Need to work harder!"
else:
    grade = "F"
    feedback = "Fail. Please study more!"

print(f"Average Marks: {average}")
print(f"Grade: {grade}")
print(f"Feedback: {feedback}")

print("\n" + "=" * 60)
print("Subject-wise Performance Analysis")
print("=" * 60)

subjects = ["Math", "Science", "English", "History"]
subject_marks = dict(zip(subjects, marks))

print("\nDetailed Report:")
for subject, mark in subject_marks.items():
    percentage = (mark / 100) * 100
    
    if mark >= 90:
        status = "Excellent "
    elif mark >= 75:
        status = "Very Good "
    elif mark >= 60:
        status = "Good "
    else:
        status = "Need Improvement "
    
    print(f"  {subject}: {mark}/100 - {status}")

# Find best and worst subjects
best_subject = max(subject_marks, key=subject_marks.get)
worst_subject = min(subject_marks, key=subject_marks.get)

print(f"\n Best Subject: {best_subject} ({subject_marks[best_subject]} marks)")
print(f"  Weakest Subject: {worst_subject} ({subject_marks[worst_subject]} marks)")
print(f"   Focus more on {worst_subject}!")


print("\n" + "=" * 60)
print("Real-Life Application: Progress Tracking")
print("=" * 60)


previous_marks = [80, 85, 55, 70]
current_marks = [85, 90, 60, 75]

print("\nSemester Comparison:")
print(f"Previous Average: {sum(previous_marks) / len(previous_marks)}")
print(f"Current Average: {sum(current_marks) / len(current_marks)}")

improvement = sum(current_marks) - sum(previous_marks)
if improvement > 0:
    print(f"âœ“ Improvement: +{improvement} marks! Great job! ")
elif improvement < 0:
    print(f"âœ— Decline: {improvement} marks. Need to work harder!")
else:
    print("= Same performance. Try to improve!")


print("\nSubject-wise Improvement:")
for i, subject in enumerate(subjects):
    change = current_marks[i] - previous_marks[i]
    if change > 0:
        print(f"  â†‘ {subject}: +{change} marks (Improved!)")
    elif change < 0:
        print(f"  â†“ {subject}: {change} marks (Declined)")
    else:
        print(f"  â†’ {subject}: No change")


print("\n" + "=" * 60)
print("Target Calculator")
print("=" * 60)

target_average = 80
current_total = sum(marks)
subjects_completed = len(marks)
remaining_subjects = 1  # Assume 1 more exam

required_total = target_average * (subjects_completed + remaining_subjects)
required_marks = required_total - current_total

print(f"Current Average: {average}")
print(f"Target Average: {target_average}")
print(f"Subjects completed: {subjects_completed}")
print(f"Remaining subjects: {remaining_subjects}")

if required_marks <= 100:
    print(f"\nTo achieve {target_average}% average:")
    print(f"  You need to score {required_marks} marks in next exam")
    if required_marks >= 90:
        print("   You need to work really hard!")
    elif required_marks >= 75:
        print("   Good effort needed!")
    else:
        print("  âœ“ Easily achievable! Keep studying!")
else:
    print(f"\n  Target of {target_average}% is not achievable anymore")
    realistic_target = (current_total + 100) / (subjects_completed + remaining_subjects)
    print(f"  Realistic target: {realistic_target:.1f}%")
```
