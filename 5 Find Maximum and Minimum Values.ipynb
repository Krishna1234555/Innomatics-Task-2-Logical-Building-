```python
attendance = ["P", "P", "A", "P", "P"]
present_days = attendance.count("P")
absent_days = attendance.count("A")
total_days = len(attendance)

attendance_percentage = (present_days / total_days) * 100

print(f"Attendance record: {attendance}")
print(f"Present days: {present_days}")
print(f"Absent days: {absent_days}")
print(f"Total days: {total_days}")
print(f"Attendance Percentage: {attendance_percentage}%")

print("\n--- Eligibility Check ---")
required_percentage = 75

if attendance_percentage >= required_percentage:
    print(f"✓ Good attendance! ({attendance_percentage}% >= {required_percentage}%)")
    print("  You're eligible for exams.")
else:
    print(f"✗ Low attendance! ({attendance_percentage}% < {required_percentage}%)")
    print("  Need to improve to be eligible for exams.")
present_count = 0
for day in attendance:
    if day == "P":
        present_count += 1

manual_percentage = (present_count / len(attendance)) * 100
print(f"Attendance Percentage (manual): {manual_percentage}%")

print("\n" + "=" * 60)
print("Real-Life Application: Weekly Attendance Report")
print("=" * 60)
weekly_attendance = {
    "Monday": "P",
    "Tuesday": "P",
    "Wednesday": "A",
    "Thursday": "P",
    "Friday": "P",
    "Saturday": "P"
}

print("Day-wise attendance:")
present_count = 0
for day, status in weekly_attendance.items():
    symbol = "✓" if status == "P" else "✗"
    status_text = "Present" if status == "P" else "Absent"
    print(f"  {symbol} {day}: {status_text}")
    if status == "P":
        present_count += 1

week_percentage = (present_count / len(weekly_attendance)) * 100
print(f"\nWeekly Attendance: {week_percentage}%")
print(f"Present: {present_count}/{len(weekly_attendance)} days")
```
