```python
employees = {"Ram": 75000, "Anita": 68000, "Kiran": 72000}

print("\n--- Method 1: Using 'in' Keyword ---")
search_name = "Ram"

if search_name in employees:
    print(f" {search_name} exists in employee database")
    print(f"  Salary: ₹{employees[search_name]}")
else:
    print(f" {search_name} not found in employee database")

# Searching for someone who doesn't exist
search_name2 = "John"
if search_name2 in employees:
    print(f" {search_name2} exists")
else:
    print(f"{search_name2} not found in employee database")


print("\n--- Method 2: Using Get Method ---")
salary = employees.get("Ram")
if salary:
    print(f"Ram's salary: ₹{salary}")

salary2 = employees.get("John", "Not Found")
print(f"John's salary: {salary2}")

print("\n--- Method 3: Using Keys Method ---")
if "Anita" in employees.keys():
    print(" Anita exists in employee database")

print("\n--- Method 4: Using Try-Except ---")
try:
    salary = employees["Kiran"]
    print(f" Kiran found! Salary: ₹{salary}")
except KeyError:
    print(" Employee not found")

try:
    salary = employees["Sarah"]
    print(f" Sarah found! Salary: ₹{salary}")
except KeyError:
    print("✗ Sarah not found in database")

# Check multiple employees
print("\n" + "=" * 60)
print("Checking Multiple Employees")
print("=" * 60)
names_to_check = ["Ram", "John", "Anita", "Sarah", "Kiran"]

print("\nSearch Results:")
for name in names_to_check:
    if name in employees:
        print(f"  {name} - Found (Salary: ₹{employees[name]:,})")
    else:
        print(f"  {name} - Not Found")


print("\n" + "=" * 60)
print("Real-Life Application: User Login System")
print("=" * 60)

user_database = {
    "ram@email.com": "pass123",
    "anita@email.com": "secure456",
    "kiran@email.com": "mypass789"
}

login_email = "ram@email.com"
login_password = "pass123"

print(f"Login attempt: {login_email}")

if login_email in user_database:
    if user_database[login_email] == login_password:
        print("Login successful! Welcome!")
    else:
        print(" Incorrect password")
else:
    print(" Email not found. Please register.")

# Another login attempt
login_email2 = "john@email.com"
print(f"\nLogin attempt: {login_email2}")

if login_email2 in user_database:
    print(" User found")
else:
    print("User not found. Would you like to register?")

# Real-life application: Product Inventory Check
print("\n" + "=" * 60)
print("Real-Life Application: Product Inventory")
print("=" * 60)

inventory = {
    "Laptop": 15,
    "Mouse": 50,
    "Keyboard": 30,
    "Monitor": 20
}

# Check product availability
product = "Laptop"
if product in inventory:
    stock = inventory[product]
    if stock > 0:
        print(f" {product} is available!")
        print(f"  Stock: {stock} units")
    else:
        print(f" {product} is out of stock")
else:
    print(f" {product} not found in inventory")

# Check another product
product2 = "Tablet"
if product2 in inventory:
    print(f" {product2} available: {inventory[product2]} units")
else:
    print(f" {product2} not available in inventory")

# Show all available products
print("\n--- All Available Products ---")
for product, quantity in inventory.items():
    status = "In Stock" if quantity > 0 else "Out of Stock"
    print(f"  • {product}: {quantity} units ({status})")
```
