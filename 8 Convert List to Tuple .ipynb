```python
my_list = [10, 20, 30]
print("\n--- Converting List to Tuple ---")
my_tuple = tuple(my_list)

print(f"Original List: {my_list}")
print(f"Type: {type(my_list)}")
print(f"\nConverted Tuple: {my_tuple}")
print(f"Type: {type(my_tuple)}")


print("\n" + "=" * 60)
print("Key Differences Between List and Tuple")
print("=" * 60)

print("\n1. MUTABILITY (Can it be changed?)")
print("-" * 40)


test_list = [1, 2, 3]
print(f"Original list: {test_list}")
test_list.append(4)
print(f"After adding 4: {test_list}")
test_list[0] = 100
print(f"After changing first item: {test_list}")


test_tuple = (1, 2, 3)
print(f"\nOriginal tuple: {test_tuple}")
print("Trying to change tuple...")
try:
    test_tuple[0] = 100
except TypeError as e:
    print(f" Error: {e}")
    print("Tuples cannot be modified!")

print("\n2. SYNTAX")
print("-" * 40)
print(f"List uses square brackets: {my_list}")
print(f"Tuple uses parentheses: {my_tuple}")

print("\n3. SPEED")
print("-" * 40)
import sys
print(f"List size in memory: {sys.getsizeof(my_list)} bytes")
print(f"Tuple size in memory: {sys.getsizeof(my_tuple)} bytes")
print("✓ Tuples are faster and use less memory!")


print("\n" + "=" * 60)
print("When to Use List vs Tuple?")
print("=" * 60)

print("\n✓ USE LIST when:")
print("  • You need to add/remove items")
print("  • Data will change over time")
print("  • Example: Shopping cart, to-do list")

print("\n✓ USE TUPLE when:")
print("  • Data should never change")
print("  • You want to protect data from modification")
print("  • Example: Coordinates, dates, configuration")


print("\n" + "=" * 60)
print("Real-Life Applications")
print("=" * 60)

print("\n1. GPS Coordinates (TUPLE - Should not change)")
location = (19.0760, 72.8777) 
print(f"Mumbai Location: {location}")
print("  These coordinates are fixed and permanent!")

print("\n2. Student Marks (LIST - Can change if updated)")
marks = [85, 90, 78]
print(f"Marks: {marks}")
marks.append(92)  
print(f"After new exam: {marks}")
print("  Marks can be updated as new exams happen!")

print("\n3. Date of Birth (TUPLE - Never changes)")
dob = (15, 8, 1947)  
print(f"Independence Day: {dob}")
print("  This date is permanent!")

print("\n4. Database Record (TUPLE - Fixed structure)")
employee_record = (101, "Ravi", "Engineer", 75000)
print(f"Employee: {employee_record}")
print("  ID, Name, Position, Salary - structured data!")


print("\n" + "=" * 60)
print("Converting Back: Tuple to List")
print("=" * 60)
original_tuple = (1, 2, 3, 4, 5)
converted_list = list(original_tuple)
print(f"Tuple: {original_tuple}")
print(f"Converted to List: {converted_list}")
print("Now you can modify it!")
converted_list.append(6)
print(f"After modification: {converted_list}")
```
